```python
import pandas as pd
```

```python
simpsons_episodes = pd.read_pickle("./simpsons_episodes.pkl")
```

```python
simpsons_episodes.shape
```

```python
def remove_str(votes):
    for r in ((',',''), ('(',''),(')','')):
        votes = votes.replace(*r)
        
    return votes
```

```python
simpsons_episodes['total_votes'] = simpsons_episodes.total_votes.apply(remove_str).astype(int)
simpsons_episodes['rating'] = simpsons_episodes.rating.astype(float)
simpsons_episodes['airdate'] = pd.to_datetime(simpsons_episodes.airdate)
```
